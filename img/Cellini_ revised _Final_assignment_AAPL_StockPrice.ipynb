```python
# Import sklearn.model, prepare for Train/test split for linear regression
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=30)  
```

```python
# Import Linear Regression from sklearn
from sklearn.linear_model import LinearRegression
```

```python
ols = LinearRegression()
ols.fit(X_train, y_train)
```

```python
# Print Coeficient, intersect and R-square
print("Intercept is:" + str(ols.intercept_))
print("The set of coeficient are:" + str(ols.coef_))
print("The R-Square value is:" + str(ols.score(X_train,y_train)))
```

```python
# Predict the dependent y from X_test
y_pred = ols.predict(X_test)
```

```python
# Import library for Mean Absolute Error, Mean Square Error and RMSE
from sklearn.metrics import mean_squared_error, mean_absolute_error 
import math
```

```python
print("Mean Absolute Error: ", mean_absolute_error(y_test, y_pred))
print("Mean Squared Error: ", mean_squared_error(y_test, y_pred))
print("RMSE : ", math.sqrt(mean_squared_error(y_test, y_pred)))
```

```python
# Calculate the performance
performance = pd.DataFrame({'PREDICTIONS':y_pred,'ACTUAL VALUES':y_test})
performance.head()
```

```python
# Calculate the error by compare the actual values with the predicted value
performance['error']=performance['ACTUAL VALUES'] - performance['PREDICTIONS']
performance.head()
```

```python
# Rest Index 
performance.reset_index(drop=True,inplace=True)
performance.reset_index(inplace=True)
performance.head()
```

```python
# Plot the Residuals (error) from observations
fig = plt.figure(figsize=(10,5))
plt.bar('index','error',data=performance, color='grey', width=0.3)
plt.xlabel("Observations")
plt.ylabel("Residuals")
plt.show()
```

```python
# Plot the distribution of residuals
sns.displot(performance['error'], bins=12, kde='True')
```

```python
# Use Normal Probability Plot (Q-Q plot) to indicate if the residuals should align the straight line
import pylab
import scipy.stats as stats
stats.probplot(performance['error'], dist="norm", plot=pylab)
pylab.show()
```
