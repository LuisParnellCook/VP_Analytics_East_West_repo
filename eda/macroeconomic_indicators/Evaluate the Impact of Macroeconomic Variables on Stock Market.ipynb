```python
# Import library matplotlib and seaborn
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Draws a scatterplot and fits a linear regression mode
sns.lmplot(x = "Unemployment Rate",
           y = "AAPL Stock Closing Price",
           data = aapl_compare,
           scatter_kws={'alpha':0.3})
```

```python
# Draws a scatterplot and fits a linear regression mode
sns.lmplot(x = "GDP Growth",
           y = "AAPL Stock Closing Price",
           data = aapl_compare,
           scatter_kws={'alpha':0.3})
```

```python
# Draws a scatterplot and fits a linear regression mode
sns.lmplot(x = "Interest Rate",
           y = "AAPL Stock Closing Price",
           data = aapl_compare,
           scatter_kws={'alpha':0.3})
```

```python
# Draws a scatterplot and fits a linear regression mode
sns.lmplot(x = "Inflation Rate",
           y = "AAPL Stock Closing Price",
           data = aapl_compare,
           scatter_kws={'alpha':0.3})
```

```python
# Import sklearn.model, prepare for Train/test split for linear regression
from sklearn.model_selection import train_test_split
```

```python
X_train_aapl_sp, X_test_aapl_sp, y_train_aapl_sp, y_test_aapl_sp = train_test_split(X_aapl_sp, y_aapl_sp, test_size=0.4, random_state=30)  
```

```python
# Import Linear Regression from sklearn
from sklearn.linear_model import LinearRegression
```

```python
ols = LinearRegression()
ols.fit(X_train_aapl_sp, y_train_aapl_sp)
```

```python
# Print Coeficient, intersect and R-square
print("Intercept is:" + str(ols.intercept_))
print("The set of coeficient are:" + str(ols.coef_))
print("The R-Square value is:" + str(ols.score(X_train_aapl_sp,y_train_aapl_sp)))
```

```python
# Predict the dependent y from X_test
y_pred_aapl_sp = ols.predict(X_test_aapl_sp)
```

```python
# Import library for Mean Absolute Error, Mean Square Error and RMSE
from sklearn.metrics import mean_squared_error, mean_absolute_error 
import math
```

```python
print("Mean Absolute Error (AAPL Stock Closing Price): ", mean_absolute_error(y_test_aapl_sp, y_pred_aapl_sp))
print("Mean Squared Error (AAPL Stock Closing Price): ", mean_squared_error(y_test_aapl_sp, y_pred_aapl_sp))
print("RMSE (AAPL Stock Closing Price): ", math.sqrt(mean_squared_error(y_test_aapl_sp, y_pred_aapl_sp)))
```

```python
# Calculate the performance
performance_aapl_sp = pd.DataFrame({'PREDICTIONS (AAPL STOCK CLOSING PRICE)':y_pred_aapl_sp,'ACTUAL VALUES (AAPL STOCK CLOSING PRICE)':y_test_aapl_sp})
performance_aapl_sp.head()
```

```python
# Calculate the error by compare the actual values with the predicted value
performance_aapl_sp['aapl_sp_error']=performance_aapl_sp['ACTUAL VALUES (AAPL STOCK CLOSING PRICE)'] - performance_aapl_sp['PREDICTIONS (AAPL STOCK CLOSING PRICE)']
performance_aapl_sp.head()
```

```python
# Rest Index 
performance_aapl_sp.reset_index(drop=True,inplace=True)
performance_aapl_sp.reset_index(inplace=True)
performance_aapl_sp.head()
```

```python
# Plot the Residuals (error) from observations
fig = plt.figure(figsize=(10,5))
plt.bar('index','aapl_sp_error',data=performance_aapl_sp, color='grey', width=0.3)
plt.xlabel("Observations (AAPL Stock Closing Price)")
plt.ylabel("Residuals (AAPL Stock Closing Price)")
plt.show()
```

```python
# Plot the distribution of residuals
sns.displot(performance_aapl_sp['aapl_sp_error'], bins=12, kde='True')
```

```python
# Use Normal Probability Plot (Q-Q plot) to indicate if the residuals should align the straight line
import pylab
import scipy.stats as stats
stats.probplot(performance_aapl_sp['aapl_sp_error'], dist="norm", plot=pylab)
pylab.show()
```
